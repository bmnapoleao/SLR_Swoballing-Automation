```python
req.content
```

```python
data = BeautifulSoup(req.text,'html.parser')
data
```

```python
req.text.find('references')!=-1 # to find presence of ref in webpage
```

```python
refer = data.find_all("span", class_="references__note")[2]
```

```python
refer.text
```

```python
for s in refer.stripped_strings:
    print(s)
```

#### Rendering Javascript pages

IEEE, Scopus: Springer, Google Scholar

```python
from requests_html import HTMLSession
```

```python
url = 'https://ieeexplore.ieee.org/document/7750588'
'''
https://ieeexplore.ieee.org/document/7750588
https://link.springer.com/article/10.1007/s11696-022-02290-1
https://www.semanticscholar.org/paper/Secondary-Studies-in-the-Academic-Context%3A-A-and-Felizardo-Souza/c4ed69120d4169f292559316c687db17d276d9dd
'''
```

```python
s = HTMLSession()
r = s.get(url)
```

```python
r.html.render(sleep = 1)
```
